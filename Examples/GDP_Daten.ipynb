```python
#This example script will crawl the GDP Planetarien Liste and save all information in a new file

import requests
import urllib.request
from bs4 import BeautifulSoup
#these libs are important for the crawling process
```

```python
urls = ['https://www.gdp-planetarium.org/planetarien/liste-der-planetarien/d-ost',
        'https://www.gdp-planetarium.org/planetarien/liste-der-planetarien/d-nord',
        'https://www.gdp-planetarium.org/planetarien/liste-der-planetarien/d-mitte',
        'https://www.gdp-planetarium.org/planetarien/liste-der-planetarien/d-sued']
#currently, all places are saved in these four links
```

```python
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    sternwarten = soup.findAll('h3')
    sternwarten_data = soup.findAll('p')
    final = []
    for i in range(0, len(sternwarten)):
        data = str(sternwarten_data[3*i])[3:-4] + '<br/>' + str(sternwarten_data[3*i+1])[3:-4] + '<br/>' + str(sternwarten_data[3*i+2])[3:-4]
        data = data.split('<br/>')
        final.append((str(sternwarten[i])[4:-5], data))

fil = open('gdp.data', 'w')
for data in final:
    print(*data, sep='\t', file=fil)
fil.close()
#reads the url in and parses the html text, h3 is marker for the city, p for the information
#I don't need the first three and last 4 strings (ignore them)
#split everything with <br/>
#save in a final form, can be converted and edited later
```
